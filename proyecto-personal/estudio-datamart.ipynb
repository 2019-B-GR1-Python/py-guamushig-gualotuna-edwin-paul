````python
import pandas as pd
import os
import xlsxwriter
import matplotlib.pyplot as plt
import seaborn as sns

path_csv = './dataframe/WorldCupMatches.csv'
path_csv2 = './dataframe/WorldCups.csv'
path_guardar_pickle = './pickle/WorldCupMatches.pickle'


df = pd.read_csv(path_csv, nrows = 852)
df2 = pd.read_csv(path_csv2, nrows = 852)
````

### Obtendremos el numero de asistentes a cada mundial

Una vez que se haya obtenido el numero de asistentes a cada copa del mundo, se debera mostrar en una grafica de barras:

````python
total_asistentes_por_mundial = df.groupby('Year')['Attendance'].sum().reset_index()
total_asistentes_por_mundial = total_asistentes_por_mundial.astype(int)
total_asistentes_por_mundial.columns = ['Anio','Asistentes']
total_asistentes_por_mundial
````

````python
plt.figure(figsize=(18,7))
sns.barplot(total_asistentes_por_mundial['Anio'],total_asistentes_por_mundial['Asistentes'])
plt.grid(True)
plt.title('Asistentes por anio', color='g', size = 25)
plt.show()
````

### Las 5 ciudades que han tenido mas asistencia de publico

````python
ciudades_asistencia =  df.groupby('City')['Attendance'].sum().reset_index()
ciudades_asistencia.columns = ['Ciudad','Asistentes']
ciudades_asistencia = ciudades_asistencia.sort_values(by = 'Asistentes', ascending = False)
ciudades_asistencia = ciudades_asistencia[:5]
ciudades_asistencia
````

````python
plt.figure(figsize=(18,7))
sns.barplot(y = ciudades_asistencia['Ciudad'], x = ciudades_asistencia['Asistentes'])
plt.grid(True)
plt.title('Top ciudades mas visitadas', color='r', size = 25)
plt.show()
````

### Selecciones que han jugado mas partidos.

````python
partidos_local = df["Home Team Name"].value_counts().reset_index()
partidos_local.columns = ["Pais","Partidos"]
partidos_visita = df["Away Team Name"].value_counts().reset_index()
partidos_visita.columns = ["Pais","Partidos"]
partidos_jugados = pd.concat([partidos_local,partidos_visita],axis=0)
partidos_jugados = partidos_jugados.groupby("Pais")["Partidos"].sum().reset_index().sort_values(by="Partidos",ascending=False)
partidos_jugados = partidos_jugados[:10]
partidos_jugados
````

````python
plt.figure(figsize=(18,7))
plt.plot(partidos_jugados['Pais'], partidos_jugados['Partidos'],  linestyle='-', marker='o',  markerfacecolor='red')
plt.grid(True)
plt.xlabel('Paises')
plt.ylabel('Goles') 
plt.title('Paises con mas partidos jugados', color='b', size = 25)
plt.show()
````

### Goles por mundial

````python
goles_local = df.groupby('Year')['Home Team Goals'].sum().reset_index()
goles_local.columns = ['Anio','Goles']
goles_visita = df.groupby('Year')['Away Team Goals'].sum().reset_index()
goles_visita.columns = ['Anio','Goles']
goles_anotados = pd.concat([goles_local, goles_visita],axis=0)
goles_anotados = goles_anotados.groupby('Anio')['Goles'].sum().reset_index()
goles_anotados
````

````python
plt.figure(figsize=(20,7))
plt.grid(True)
plt.plot(goles_anotados['Anio'],goles_anotados['Goles'] , color='y', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='g', markersize=12) 
plt.xlabel('Anio')
plt.ylabel('Goles')
plt.title('Goles por mundial', color='g', size = 25)
plt.show()
````

### Ciudades con menos partidos jugados

````python
ciudades_partidos = df['City'].value_counts().reset_index()
ciudades_partidos.columns = ['Ciudad', 'Partidos']
ciudades_partidos = ciudades_partidos[-20:]
ciudades_partidos
````

````python
plt.figure(figsize=(20,7))
plt.grid(True)
plt.scatter(ciudades_partidos['Ciudad'], ciudades_partidos['Partidos'])
plt.xlabel('Ciudad')
plt.ylabel('Partidos')
plt.title('Partidos por ciudad', color='b', size = 25)
plt.show()
````

### Promedio de asistencia de los ultimos 10 mundiales

````python
promedio_asistencia = df.groupby("Year")["Attendance"].mean().reset_index()
promedio_asistencia["Year"] = promedio_asistencia["Year"].astype(int)
promedio_asistencia.columns = ['Anio', 'Promedio']
promedio_asistencia = promedio_asistencia[-10:]
promedio_asistencia
````

````python
plt.figure(figsize=(20,7))
plt.grid(True)
plt.plot(promedio_asistencia['Anio'], promedio_asistencia['Promedio'], 'r^')
plt.title('Promedio de asistencia', color='b', size = 25)
plt.xlabel('Anio')
plt.ylabel('Promedio')
plt.show()
````

### Selecciones campeonas

````python
campeones = df2.groupby('Winner')['Year'].count().reset_index().sort_values(by="Year",ascending=False)
campeones.columns = ['Campeon', 'Cantidad']   

campeones
````

````python
labels = campeones['Campeon']
labels.to_string()
fig1, ax1 = plt.subplots()
ax1.pie(campeones['Cantidad'], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()
````

### Selecciones que mas veces han participado

````python
selecciones_local = df[['Year', 'Home Team Name']]
selecciones_local.columns = ['Anio', 'Seleccion']
selecciones_visita = df[['Year', 'Home Team Name']]
selecciones_visita.columns = ['Anio', 'Seleccion']
participaciones = pd.concat([selecciones_local, selecciones_visita], axis = 0)
participaciones_anio = participaciones.groupby(['Anio', 'Seleccion']).count().reset_index()
participaciones_anio = participaciones_anio['Seleccion'].value_counts().reset_index()
participaciones_anio = participaciones_anio[:7]
participaciones_anio.columns = ['Pais', 'Participaciones']
participaciones_anio
````

````python
fig, ax = plt.subplots()
ax.plot(participaciones_anio['Participaciones'], '-o', ms=20, alpha=0.7, mfc='orange')
ax.grid()
plt.xlabel('Participaciones')
plt.ylabel('Pais')
plt.show()
````

### Partidos con mas goles marcados

````python
df["Goles totales"] = df["Home Team Goals"] + df["Away Team Goals"]
goles_mundial = df[['Year', 'Goles totales']]
goles_mundial
````

````python
plt.figure(figsize=(20,7))
sns.boxplot(y=goles_mundial["Goles totales"],
            x=goles_mundial["Year"])
plt.grid(True)
plt.title("Goles marcados por anio",color='b')
plt.show()
````

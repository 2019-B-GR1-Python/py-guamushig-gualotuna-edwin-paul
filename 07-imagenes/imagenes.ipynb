```python
import numpy as np
from scipy import misc
```

```python
mapache = misc.face()
mapache
```

```python
type(mapache)
mapache.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
plt.imshow(mapache)
print(mapache[0][0][1])
print(mapache[0][0][2])
plt.show()
```

```python
# extraer una parte del mapache

mapache_parcial = mapache[385:,512:,:]
plt.imshow(mapache_parcial)
plt.show()
```

```python
# extraer mitad izquierda del mapache

mapache_parte_uno,mapache_parte_dos =np.split(mapache,2,axis=1)
plt.imshow(mapache_parte_uno)
plt.show()
```

```python
# extraer mitad derecha del mapache

plt.imshow(mapache_parte_dos)
plt.show()
```

```python
plt.imshow(mapache[::-1])  
plt.show()
```

```python
mapache_parcial = mapache[400:,600:,:]
plt.imshow(mapache_parcial)
plt.show()
```

```python
def div_mapache(arreglo):
    mapache_izquierda,mapache_derecha =np.split(arreglo,2,axis=1)
    plt.imshow(mapache_izquierda)
    plt.show()
    plt.imshow(mapache_derecha)
    plt.show()
```

```python
div_mapache(mapache)
```
